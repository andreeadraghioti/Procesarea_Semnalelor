```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def function1(n1, n2):
    return np.sin(2 * np.pi * n1 + 3 * np.pi * n2)

def function2(n1, n2):
    return np.sin(4 * np.pi * n1) + np.cos(6 * np.pi * n2)

def function3(N, m1, m2):
    if m1 == 0.5 and m2 == N - 5:
        return 1
    else:
        return 0

def function4(N, m1, m2):
    if m1 == 5 and m2 == 0:
        return 1
    else:
        return 0

def function5(N, m1, m2):
    if m1 == 5 and m2 == 5:
        return 1
    else:
        return 0
    
n_values = np.linspace(0, 10, 100)

n1, n2 = np.meshgrid(n_values, n_values)
x_values1 = function1(n1, n2)
x_values2 = function2(n1, n2)
x_values3 = np.vectorize(function3)(n_values.shape[0], n1, n2)
x_values4 = np.vectorize(function4)(n_values.shape[0], n1, n2)
x_values5 = np.vectorize(function5)(n_values.shape[0], n1, n2)

plt.figure(figsize=(15, 5))
plt.imshow(x_values1, extent=(0, 10, 0, 10), origin='lower', cmap='viridis')
plt.colorbar()
plt.title('sin(2πn1 + 3πn2)')

plt.figure(figsize=(15, 5))
plt.imshow(x_values2, extent=(0, 10, 0, 10), origin='lower', cmap='viridis')
plt.colorbar()
plt.title('sin(4πn1) + cos(6πn2)')

plt.figure(figsize=(15, 5))
plt.imshow(x_values3, extent=(0, 10, 0, 10), origin='lower', cmap='viridis')
plt.colorbar()
plt.title('Y0.5, N-5')

plt.figure(figsize=(15, 5))
plt.imshow(x_values4, extent=(0, 10, 0, 10), origin='lower', cmap='viridis')
plt.colorbar()
plt.title('Y5, 0')

plt.figure(figsize=(15, 5))
plt.imshow(x_values5, extent=(0, 10, 0, 10), origin='lower', cmap='viridis')
plt.colorbar()
plt.title('Y5, 5')

plt.show()
```
