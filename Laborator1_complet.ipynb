```python
#laborator1
#exercitiul 1
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0, 0.03, 0.0005)

semnal_x = np.cos(520 * np.pi * t + np.pi/3)
semnal_y = np.cos(280 * np.pi * t - np.pi/3)
semnal_z = np.cos(120 * np.pi * t + np.pi/3)

fig, axs = plt.subplots(3)
fig.suptitle('Semnale x(t), y(t), z(t)')
axs[0].plot(t,semnal_x)
axs[1].plot(t,semnal_y)
axs[2].plot(t,semnal_z)

fs = 200
n = np.arange(0, len(t))
semnal_discret_x = np.cos(520 * np.pi * n * 1/fs + np.pi/3)
semnal_discret_y = np.cos(280 * np.pi * n * 1/fs - np.pi/3)
semnal_discret_z = np.cos(120 * np.pi * n * 1/fs + np.pi/3)

fig2, axs2 = plt.subplots(3)
fig2.suptitle('Semnale discrete x[n], y[n], z[n]')
axs2[0].plot(n,semnal_discret_x)
axs2[1].plot(n,semnal_discret_y)
axs2[2].plot(n,semnal_discret_z)


plt.show()
```

```python
#exercitiul2
import numpy as np
import matplotlib.pyplot as plt

t_a = np.linspace(0, 1, 1600)
semnal_a = np.sin(2 * np.pi * 400 * t_a)

t_b = np.linspace(0, 3, 800 * 3)
semnal_b = np.sin(2 * np.pi * 800 * t_b)

t_c = np.linspace(0, 1, 240)
semnal_c = 2 * (t_c - np.floor(t_c + 0.5))

t_d = np.linspace(0, 1, 300)
semnal_d = np.sign(np.sin(2 * np.pi * 300 * t_d))

fig, axs = plt.subplots(4)
fig.suptitle('Exercitiul 1')
axs[0].plot(t_a,semnal_a)
axs[1].plot(t_b,semnal_b)
axs[2].plot(t_c,semnal_c)
axs[3].plot(t_d,semnal_d)


semnal_e = np.random.rand(128, 128)

plt.figure(figsize=(6, 6))
plt.imshow(semnal_e, cmap='coolwarm')
plt.title('Semnal 2D aleator')

semnal_f = np.ones((128, 128))
semnal_f[48:80, 48:80] = 0

plt.figure(figsize=(6, 6))
plt.imshow(semnal_f, cmap='bone')
plt.title('Semnal 2D la alegere')

plt.show()
```
