```python
import numpy as np
import matplotlib.pyplot as plt
#ex1
n = 8
dft1 = np.zeros((n,n),dtype=complex)
for m in range(n):
  for k in range(n):
    dft1[m][k] = np.e**(2j*np.pi*m*k/8)

fig,axs = plt.subplots(8)
for item in range (8):
  axs[item].plot(dft1[item].real)
  axs[item].plot(dft1[item].imag)

plt.show()

este_unitara = np.linalg.norm(np.abs(np.dot(np.conjugate(dft1.T),dft1))-n*np.identity(n),2)
if este_unitara <= 10**(-10):
  print("este unitara")
else:
  print("nu este unitara")


t = np.linspace(0,1500,8)
dft2 = np.matmul(dft1,t)

```
