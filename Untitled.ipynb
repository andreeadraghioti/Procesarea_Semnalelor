```python
import numpy as np
import matplotlib.pyplot as plt

def sample_bivariate_gaussian(mean, covariance, num_samples=1000):
    eigvals, eigvecs = np.linalg.eigh(covariance)
    n = np.random.randn(num_samples, 2)
    samples = n @ np.sqrt(np.diag(eigvals)) @ eigvecs.T + mean
    
    plt.figure(figsize=(10, 6))
    plt.scatter(samples[:, 0], samples[:, 1], alpha=0.6)
    plt.title('Scatter Plot of Bivariate Gaussian Distribution')
    plt.xlabel('X value')
    plt.ylabel('Y value')
    plt.grid(True)
    plt.axis('equal')
    plt.show()

mean_bi = np.array([0, 0])
covariance_bi_specific = np.array([[1, 3/5], [3/5, 2]])

sample_bivariate_gaussian(mean_bi, covariance_bi_specific)
```
