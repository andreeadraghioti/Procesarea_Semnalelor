```python
#laboratorul 8
#a
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

N = 1000

t = np.arange(N)

trend = 0.02 * t**2

season = 10 * np.sin(0.02 * np.pi * t) + 5 * np.cos(0.04 * np.pi * t)

noise = np.random.normal(0, 1, N)

time_series = trend + season + noise

plt.figure(figsize=(12, 6))
plt.subplot(4, 1, 1)
plt.plot(t, time_series, label='Seria de timp')
plt.legend()

plt.subplot(4, 1, 2)
plt.plot(t, trend, label='Trend')
plt.legend()

plt.subplot(4, 1, 3)
plt.plot(t, season, label='Sezon')
plt.legend()

plt.subplot(4, 1, 4)
plt.plot(t, noise, label='Zgomot')
plt.legend()

plt.tight_layout()
plt.show()

#b
from statsmodels.graphics.tsaplots import plot_acf

autocorrelation = plot_acf(time_series, lags=50)
plt.title('Autocorelaie')
plt.show()
```
